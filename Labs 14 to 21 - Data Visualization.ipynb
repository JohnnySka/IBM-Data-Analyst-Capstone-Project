Import pandas module.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
QUERY = """
SELECT ConvertedComp FROM master
"""
df = pd.read_sql_query(QUERY, conn)

sns.set_style(style="darkgrid")
plt.figure(figsize=(8,6))
sns.histplot(data=df, x="ConvertedComp", bins=50, color="red")
plt.title("Histogram of ConvertedComp Column", fontsize=15)
plt.xlabel("ConvertedComp", fontsize=12)
plt.ylabel("Counts", fontsize=12)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(15,8))
sns.boxplot(data=df, x="Age", color="green")
plt.title("Boxplot of Age Distribution", fontsize=22)
plt.xlabel("Age", fontsize=20)
plt.tick_params(axis="both", labelsize=16)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
QUERY = """
SELECT Age, WorkWeekHrs FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="Age", y="WorkWeekHrs", color="purple", s=100)
plt.title("Scatterplot of Age vs WorkWeekHrs", fontsize=20)
plt.xlabel("Age", fontsize=18)
plt.ylabel("WorkWeekHrs", fontsize=18)
plt.tick_params(axis="both", labelsize=14)
plt.show

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs FROM master
"""
df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x="CodeRevHrs", y="WorkWeekHrs", size="Age", alpha=0.7, sizes=(1,800))
plt.title("Bubble Plot of CodeRevHrs vs WorkWeekHrs (bubble size represents Age)", fontsize=20)
plt.xlabel("CodeRevHrs", fontsize=18)
plt.ylabel("WorkWeekHrs", fontsize=18)
plt.tick_params(axis="both", labelsize=14)
plt.show
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *, COUNT(*) AS count FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

df = pd.read_sql_query(QUERY,conn)

data = df['count']
labels = df['DatabaseDesireNextYear']

plt.figure(figsize=(15,8), facecolor="lightgrey")
palette_color = sns.color_palette('bright')
plt.pie(data, labels=labels, autopct='%1.1f%%', 
        colors=palette_color, startangle=90, explode=(0.1, 0, 0, 0, 0))
plt.title("Pie Chart of Top 5 Databases to Learn Next Year", fontsize=18)
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE Age BETWEEN 30 AND 35
"""

df = pd.read_sql_query(QUERY,conn)

WW = df['WorkWeekHrs'].median()
CR = df['CodeRevHrs'].median()


fig, ax = plt.subplots()
plt.rcParams['figure.figsize'] = [6,6]
ax.bar('Age 30 - 35', WW, label='WorkWeekHrs')
ax.bar('Age 30 - 35', CR, bottom=WW, label='CodeRevHrs')
ax.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age FROM master
WHERE Age BETWEEN 45 AND 60
"""
df = pd.read_sql_query(QUERY,conn)

medians = df.groupby('Age')['ConvertedComp'].median()

data = pd.DataFrame(list(zip(medians.index, medians.values)),
               columns = ['Age', 'MedConvertedComp'])

plt.figure(figsize=(8,6))
sns.lineplot(data=data, x='Age', y = 'MedConvertedComp', color="purple")
plt.title("Median of ConvertedComp for Ages 45-60", fontsize=15)
plt.xlabel("Age", fontsize=12)
plt.ylabel("ConvertedComp Median", fontsize=12)
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch, COUNT(*) AS Frequency FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(8,6))
sns.barplot(data = df, y = 'MainBranch', x = 'Frequency', orient = 'h')
plt.title("Bar Chart of MainBranch", fontsize=20)
plt.xlabel("Frequency", fontsize=18)
plt.ylabel("MainBranch", fontsize=18)
plt.tick_params(axis="both", labelsize=16)
plt.show()
```
